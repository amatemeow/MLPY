<h1>Алгоритм FP-Max</h1>
<h2>Выборка пользователей из Великобритании</h2>
<h3>Характерные комбинации</h3>

```python
freq_uk_fpmax = ml.fpmax(grouped_uk_bin, min_support = 0.035, use_colnames = True)
print("Найдено %d характерных комбинаций" % len(freq_uk_fpmax))
print(freq_uk_fpmax)
print("Максимальная длина набора: %d" % freq_uk_fpmax['itemsets'].apply(lambda x: len(x)).max())
```

<h3>Ассоциативные правила</h3>

```python
rules_uk_fpmax = ml.association_rules(freq_uk_fpmax, metric = "confidence", min_threshold = 0.35)
print("Найдено %d ассоциативных правил" % len(rules_uk_fpmax))
rules_uk_fpmax.sort_values("confidence", axis = 0, ascending = False)
```

<br>
<p>Из-за особенностей работы алгоритма fpmax для генерации правил не хватает информации о поддержке посылки и следствия</p>
<p>FP-Max — это вариант FP-Growth, который фокусируется на получении максимальных комбинаций. Комбинация X называется максимальныой, если X является встречается часто и не существует более частого супершаблона, содержащего X. Другими словами, частый шаблон X не может быть подшаблоном более частого шаблона, чтобы соответствовать определению максимального набора элементов. Из-за этого значения поддержки подшаблонов неизвестны, и соответственно для генерации правил не хватает информации о поддержке посылки или следствия.</p>
<p>Для получения ассоциативных правил в таком случае, нужно применить параметр "support_only=True", при этом будет вычислятся только столбец с поддержкой данного правила, для которого не требуется отсутствующая информация:</p>

```python
rules_uk_fpmax = ml.association_rules(freq_uk_fpmax, support_only=True , min_threshold = 0.05)
print("Найдено %d ассоциативных правил" % len(rules_uk_fpmax))
rules_uk_fpmax.sort_values("support", axis = 0, ascending = False)
```

<p>Как следствие, полученные ассоциативные правила отличаются от полученных при помощи других алгоритмов, однако нам неизвестно значение уверенности для этих правил. Делать выводы об актуальности данных правил приходится только на основе поддержки.</p>
